```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
WDIR = "gen-1/"
SAMPLE_NOISE_FILE = WDIR + "sample_noise.csv"
noise_file_handle = open(SAMPLE_NOISE_FILE,'r')
n_data = np.loadtxt(noise_file_handle, delimiter=',')

n_rows, NOISE_SIZE = n_data.shape
plt.title("Histogram with 'auto' bins")
_ = plt.hist(n_data, bins='auto')  # arguments are passed to np.histogram
plt.show()
```
